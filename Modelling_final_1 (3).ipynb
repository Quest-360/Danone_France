```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.model_selection import train_test_split
# Load data
data = pd.read_excel('finalads.xlsx')
# Split data into independent and dependent variables
X = data.drop(['Net Sales of Bledina','Bledina_total_emails ','app_website_visits','competitor_ooh_spends','macroeconomic_total_total_total_cci','Bledina_total_digital_catchupTV_spends','competitor_tv_spends'], axis=1)
feature_names=X.columns.to_list()
y = data['Net Sales of Bledina']
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=634422)
# Create linear regression object
reg = LinearRegression()
# Train the model using the training sets
reg.fit(X_train, y_train)
# Make predictions using the testing set
y_pred = reg.predict(X_test)
# Calculate MSE and RMSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean squared error: %.2f" % mse)
print("Root mean squared error: %.2f" % rmse)
# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print("R-squared: %.2f" % r2)
y_pred_binary = [1 if x >= 0.5 else 0 for x in y_pred]
y_test_binary = [1 if x >= 0.5 else 0 for x in y_test]
accuracy = np.mean(np.array(y_pred_binary) == np.array(y_test_binary)) * 100.0
print("Accuracy: %.2f%%" % accuracy)
```
