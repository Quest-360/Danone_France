## HISTOGRAM

#### USED TO SUMMARIZE BOTH DISCRETE AND CONTINIOUS

```python
sns.histplot(df, x='battery_capacity',bins=7,kde=True)
# kde: kernel density estimate
```

```python
sns.histplot(df, x='battery_capacity',bins=7,kde=False)
```

```python
sns.kdeplot(data = df, x='battery_capacity')
```

## PIE PLOT

#### shows categorial values in numerical values

```python
label=df['brand'].value_counts().index
count=df['brand'].value_counts().values
```

```python
label
```

```python
count
```

```python
plt.pie(count,labels=label,shadow = True,colors=['lightgreen','lightblue','maroon','beige','orange'])
```

```python
exploded=[0.05,0,0.1,0.3,0.2]
plt.pie(count,labels=label,shadow = True,colors=['lightgreen','lightblue','maroon','beige','orange'],explode=exploded)
```

```python

plt.pie(count,labels=label,shadow = False,colors=['lightgreen','lightblue','yellow','beige','orange'],autopct='%.2f%%')
```

## DOUGHNUT CHART

### Modified piechart

```python
my_circle = plt.Circle((0,0),0.35,color='white')

plt.pie(count, labels=label,autopct='%0.2f%%',colors=['red','green','blue','yellow','beige'])

p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(label,loc='upper right',title='BRANDS')
plt.show()
```

## SCATTER PLOT

### Both values should be numerical data

```python
plt.scatter(x='RAM',y='ROM',data=df)
plt.xlabel('RAM')
plt.ylabel('ROM')
plt.title('RAM VS ROM')
```

### Shows relationship between RAM and ROM

### In this case positive linear relation between RAM and ROM, ie when RAM increases ROM increases

```python
sns.scatterplot(x='RAM',y='ROM', data=df,hue='brand',style='brand')
```

# BOX PLOT

#### * Displays five number summary of a set of data (min,lower-quartile,median,higher-quartile,max)
#### * used to represent numerical data
#### * used when data has outliers
#### * also called "box & whisker diagram"

```python
plt.boxplot(df['display_size'])
plt.xlabel=('display')
plt.ylabel=('screen_size')
plt.title('box plot for display')
```

```python
sns.boxplot(y='display_size',data=df)
```

```python
sns.boxplot(x='RAM',y='processor',data=df,hue='brand')
```
