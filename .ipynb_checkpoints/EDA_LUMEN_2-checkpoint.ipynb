```python
# ID,amount.funded.by.investors : drop
# interest rate, debt to income ratio: remove % & convert to numeric
# amount.requested, open.credit.lines, revolving.credit.balance: convert to numeric
# FICO.range: replace numeric with average
# employment.length: convert to number
# loan.Length,loan.purpose,state,home.ownership: dummies
```

```python
#Dropping amount funded by investors and ID
```

```python
df = df.drop(['ID','Amount.Funded.By.Investors'], axis=1)
```

```python
df
```

```python
#tf = df['Interest.Rate'].str[:-1].astype(float)
#tf = df['Debt.To.Income.Ratio'].str[:-1].astype(float)
```

```python
#df['Interest.Rate'] = tf
```

```python
df
```

```python
for col in ['Interest.Rate','Debt.To.Income.Ratio']:
    df[col]=df[col].str.replace('%',' ')
```

```python
df
```

```python
for col in ['Amount.Requested','Interest.Rate','Debt.To.Income.Ratio']:
    df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
df.dtypes
```

```python
k=df['FICO.Range'].str.split('-',expand=True).astype(float)
df['FICO']=0.5*(k[0]+k[1])
del df['FICO.Range']
```

```python
df
```

```python
df.dtypes
```

```python
df['Employment.Length'].value_counts().to_frame()
```

```python
df['Employment.Length']=df['Employment.Length'].str.replace('years','')
df['Employment.Length']=df['Employment.Length'].str.replace('year','')
df['Employment.Length']=np.where(df['Employment.Length'].str[0:2]=='10',10,df['Employment.Length'])
df['Employment.Length']=np.where(df['Employment.Length'].str[0]=='<',1,df['Employment.Length'])
#converts object to numeric(float)#
df['Employment.Length']=pd.to_numeric(df['Employment.Length'],errors='coerce')
```

```python
q1=num_col.stroke.quantile(0.25)
q3=num_col.stroke.quantile(.75)
q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
Lower_limit =q1 - 1.5*IQR
Upper_limit =q3 + 1.5*IQR
```

```python
outlier = num_col[(num_col.stroke>Upper_limit)|(num_col.stroke<Lower_limit)]
outlier
```
