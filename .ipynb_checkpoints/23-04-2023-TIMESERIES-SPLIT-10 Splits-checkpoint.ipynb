# n_features_to_select Checker

```python
for i in range (2,45):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42,shuffle=False)

    # Create a linear regression model to use for feature selection
    lr = LinearRegression()
    lr.fit(X_train,y_train)

    # RFE features

    rfe = RFE(estimator=lr,n_features_to_select=i, step=1)

    rfe.fit(X_train,y_train)

    # Get the list of selected features
    selected_features = X.columns[rfe.support_]
    #selected_df = pd.DataFrame({'Selected Features': selected_features})

    # Print the dataframe as a table
    #selected_df

    #selected_features

    n = selected_features.tolist()
    #list(zip(X_train.columns,rfe.support_,rfe.ranking_))

    n = selected_features.tolist()

    X_train = X_train[selected_features]

    # Adding Values

    X_train['constant']=1

  
    y_train = pd.DataFrame(y_train)

    model = sm.OLS(y_train, X_train).fit()

    print(model.rsquared_adj," ",model.rsquared," ",i)
```

# Test-Train Split

# TimeSeriesSplits

```python
tscv = TimeSeriesSplit(n_splits=10)
```

```python
for train_index, test_index in tscv.split(X):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42,shuffle=False)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    lr = LinearRegression()
    
    rfe = RFE(estimator=lr,n_features_to_select=20,step=1)

    rfe.fit(X_train,y_train)

   
    selected_features = X.columns[rfe.support_]
    

    n = selected_features.tolist()
  

    X_train = X_train[selected_features]

    X_train['constant']=1

  
    y_train = pd.DataFrame(y_train)

    model = sm.OLS(y_train, X_train).fit()

    print(model.rsquared_adj," ",model.rsquared," ",train_index)
```
